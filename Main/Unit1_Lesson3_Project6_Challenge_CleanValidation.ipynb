```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Datasets/WELLCOME/WELLCOME_APCspend2013_forThinkful.csv", encoding='ISO-8859-1')
df.describe()
```

```python
import re

df['PMID'] = df['PMID/PMCID'].apply(lambda x: str((re.findall('[0-9]{8} ', str(x)))).strip('[]'))
df['PMID'] = np.where(df['PMID'] == '', None, df['PMID'].apply(lambda x: str(x).strip("' '")))

df['PMCID'] = df['PMID/PMCID'].apply(lambda x: str((re.findall('PMC[0-9]+', str(x)))).strip('[]'))
df['PMCID'] = df['PMCID'].apply(lambda x: str((re.findall('[0-9]+', x))).strip('[]'))
df['PMCID'] = np.where(df['PMCID'] == '', None, df['PMCID'].apply(lambda x: str(x).strip("''")))
```
