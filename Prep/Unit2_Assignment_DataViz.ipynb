```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
%matplotlib inline
```

```python
df = pd.read_csv('UNdata_Export_20190308_113825082.csv')
df.head()
```

# Carbon dioxide Emission of U.S.

```python
us_data = df[df['Country or Area'] == 'United States of America']

plt.plot(us_data['Year'], us_data['Value'], color='red')
plt.ylim([0, max(us_data['Value']) + 100000])
plt.ylabel('in kilotonne CO2 equivalent')
plt.title('Carbon dioxide (CO2) Emissions without Land Use, Land-Use Change and Forestry (LULUCF) of U.S.')
plt.show()
```

The chart above is a line chart. This one illustrate the development of the emision of CO2 in U.S. over time from 1990 to 2016.

```python
australia_data = df[df['Country or Area'] == 'Australia']
austria_data = df[df['Country or Area'] == 'Austria']
```

```python
plt.figure(figsize=(20, 4))

plt.subplot(1, 3, 1)
plt.plot(us_data['Year'], us_data['Value'], color='red')
plt.ylim([0, max(us_data['Value']) + 100000])
plt.ylabel('in kilotonne CO2 equivalent')
plt.title('U.S.')

plt.subplot(1, 3, 2)
plt.plot(australia_data['Year'], australia_data['Value'], color='red')
plt.ylim([0, max(australia_data['Value']) + 100000])
plt.ylabel('in kilotonne CO2 equivalent')
plt.title('Australia')

plt.subplot(1, 3, 3)
plt.plot(austria_data['Year'], austria_data['Value'], color='red')
plt.ylim([0, max(austria_data['Value']) + 100000])
plt.ylabel('in kilotonne CO2 equivalent')
plt.title('Austria')

plt.show()
```

The purpose of this chart is to have an overview about different countries and their CO2 emission development. All subplots are line charts.

# Global Emission

```python
sum_per_year = df.groupby('Year').aggregate(np.sum)
plt.plot(sum_per_year['Value'], color='red')
plt.ylim([0, max(sum_per_year['Value']) + 1000000])
plt.ylabel('in kilotonne CO2 equivalent')
plt.title('GLOBAL Carbon dioxide (CO2) Emissions without Land Use, Land-Use Change and Forestry (LULUCF)')
plt.show()
```

The chart above shows the world wide CO2 emission development over time (from 1990 to 2016). It is displayed as a
line chart.

# Emission comparison 

```python
all_data_2016 = df[df['Year'] == 2016].sort_values(by='Value',ascending=False)
```

```python
plt.figure(figsize=(20, 7))
plt.bar(all_data_2016['Country or Area'], all_data_2016['Value'])
plt.ylim([0, max(all_data_2016['Value']) + 100000])
plt.xticks(rotation=90)
plt.ylabel('in kilotonne CO2 equivalent')
plt.title('Carbon dioxide (CO2) Emissions without Land Use, Land-Use Change and Forestry (LULUCF) per Country in 2016')
plt.show()
```

The last chart is a bar chart. You can compare the CO2 emission between countries. For better comparison the countries are sorted from the highest emission to the lowest.
