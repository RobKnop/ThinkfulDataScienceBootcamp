```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def plot_hist(distribution):
    #Plot a histogram.
    plt.hist(distribution)

    # Add a vertical line at the mean.
    plt.axvline(distribution.mean(), color='b', linestyle='solid', linewidth=2)

    # Add a vertical line at one standard deviation above the mean.
    plt.axvline(distribution.mean() + distribution.std(), color='b', linestyle='dashed', linewidth=2)

    # Add a vertical line at one standard deviation below the mean.
    plt.axvline(distribution.mean() - distribution.std(), color='b', linestyle='dashed', linewidth=2) 

    # Print the histogram
    print('Mean: ', distribution.mean())
    print('Std: ', distribution.std())
    plt.show()
```

```python
#distributionvar = np.random.distributionname([arguments], 100),

n, p = 1, .5  # number of trials, probability of each trial
bernoulli = np.random.binomial(n, p, 100)

#Plot a histogram.
plot_hist(bernoulli)
```

```python
poisson = np.random.poisson(1, 100)

#Plot a histogram.
plot_hist(poisson)
```

```python
gamma = np.random.gamma(5,1, 100)

#Plot a histogram.
plot_hist(gamma)
```

```python
ngood, nbad, nsamp = 100, 2, 10
# number of good, number of bad, and number of samples
hypergeometric = np.random.hypergeometric(ngood, nbad, nsamp, 100)

#Plot a histogram.
plot_hist(hypergeometric)
```

```python
loc, scale = 10, 1
logistic = np.random.logistic(loc, scale, 1000)

plot_hist(logistic)
```

```python
mu, kappa = 0.0, 4.0 # mean and dispersion
vonmises = np.random.vonmises(mu, kappa, 100)
plot_hist(vonmises)
```

```python
normal1 = np.random.normal(5, 0.5, 100)
normal2 = np.random.normal(10, 1, 100)
added = normal1 + normal2
plot_hist(added)
```
